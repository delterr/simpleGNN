```python
import os
import time
import pickle
import numpy as np
import networkx as nx

from GNN import GNN
from tree import Node, Tree
from homomorphism import total_hom, total_hom_preprocess, refine_colours

import matplotlib.pyplot as plt
```

```python
with open("../graphs/regular/reg_200_250.pkl", "rb") as f:
    edges = pickle.load(f)
```

```python
G = GNN(edges=edges, directed=False)
A, colour_count = refine_colours(G)
```

```python
nodes = {}
trees = {}

with open("final_out.pkl", "rb") as f:
    tree_edges = pickle.load(f)

for i, arr in enumerate(tree_edges):
    for parent_val, child_val in arr:
        if parent_val not in nodes:
            nodes[parent_val] = Node(parent_val)
    
        if child_val not in nodes:
            nodes[child_val] = Node(child_val)
    
        nodes[parent_val].children.append(nodes[child_val])

    trees[i] = nodes
    nodes = {}
```

```python
deltas = {i: [] for i in range(len(trees))}

trees = list(nx.generators.nonisomorphic_trees(8))

for i in range(10):
    print(f"Processing Iteration #{i + 1}...")
    for j in range(len(trees)):
        T = Tree(edges=trees[j].edges)
        T = T.nodes[1]
        delta = time.time()
        total_hom_preprocess(T, G, colour_count, A)
        delta = time.time() - delta
        deltas[j].append(delta)
    print(f"Completed Iteration #{i + 1}!")
```
